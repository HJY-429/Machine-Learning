```python
import matplotlib.pyplot as plt
import numpy as np


#The Haar function

# we first define the heaviside function
def h(x,a): 
  return 1*(x >= a)

# we then define the step function
def u(x,s,e):
    return h(x,s)-h(x,e)

# we then define the Haar function
def haar(x):
    return u(x,-0.5,0) - u(x,0,0.5)

L=8
N_grid = 2**6 # number of points in the grid
x4 = np.linspace(-L, L, N_grid+1)
x = x4[0:N_grid]



rows_a = [1,2,4,8]
cols_b = [-2,0,2]

# Create the subplot grid
fig, axs = plt.subplots(len(rows_a), len(cols_b), figsize=(12, 8))

# Iterate through the subplots and add plots
for i in range(len(rows_a)):
    for j in range(len(cols_b)):
        a = rows_a[i]
        b = cols_b[j]
        axs[i, j].plot(x, 1 / np.sqrt(a) * haar((x - b) / a))
        axs[i, j].set_title(f"$\\Psi_{{{a},{b}}}$")  # Proper LaTeX formatting
        axs[i, j].set_ylim(-1, 1)  # Scale

# Adjust layout and show the plot
fig.tight_layout()
plt.show()

```
